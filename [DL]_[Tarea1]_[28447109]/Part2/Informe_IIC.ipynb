### Informe Parte Optimización

Pregunta #1. ¿Cuál es el impacto del parámetro ruido en el proceso de optimización? (Implementación en el notebook sumplementario)

Podemos observar:

En el Error de prueba (azul):

- Ruido 0 - 0.01: Error muy bajo, la red aprende muy bien la función original.

- Ruido 0.1: El error se mantiene bajo , la red es robusta a pequeñas perturbaciones.

- Ruido 0.5: El error se dispara, es decir, el ruido domina y la red no logra generalizar bien.

Y en el Tiempo de entrenamiento (naranja):

- Se mantiene relativamente constante, con pequeñas variaciones. Aumentando el ruido vemos que no afecta mucho el tiempo de entrenamiento en este rango.

Pregunta #2 ¿Cómo se diferencia el método del descenso del gradiente con el descenso estocástico y las versiones aceleradas?

Optimizador 1: Tiempo = 49.92s | Error Prueba = 0.005463

Optimizador 2: Tiempo = 124.27s | Error Prueba = 0.000123

Optimizador 3: Tiempo = 52.68s | Error Prueba = 0.000086

Optimizador 4: Tiempo = 56.57s | Error Prueba = 0.000024

Optimizador 5: Tiempo = 67.76s | Error Prueba = 0.000013

-  SGD clásico (Opt 1): Lento y menos preciso. Gran batch = estable pero le cuesta bajar el loss.

- SGD mini-batch (Opt 2): Con batch pequeño = converge más rápido pero es ruidoso. Tiempo total alto por cantidad de pasos, es más ineficiente.

- SGD + Momentum (Opt 3): Vemos mejora ya que converge más rápido y mejor error que SGD puro. Además que tiene bajo costo computacional.

- Adam (Opt 4): Tiene buen balance ya que baja el error rápido y mantiene un buen tiempo total.

- LBFGS (Opt 5): El más preciso y uno de los más rápidos en epochs. Requiere evaluar más dentro de cada paso (por closure), por eso su tiempo total no es el menor, pero su error es el mejor.

Pregunta #3 ¿Cómo impacta la dimensión de la red neuronal en el rendimiento del optimizador?

A medida que aumentamos la cantidad de neuronas en la red, el tiempo de entrenamiento crece considerablemente. Sin embargo, el error de prueba también se incrementa, lo que sugiere un posible sobreajuste o un uso ineficiente de la capacidad adicional. 

Podemos ver que la red más pequeña (35 neuronas) resultó ser la más eficiente, alcanzando un bajo error en menor tiempo.
